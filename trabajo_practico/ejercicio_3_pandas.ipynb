```python
df_contents = pd.read_csv("/content/drive/MyDrive/Organizacion De Datos/Trabajo_practico/contents.csv")
```

```python
df_contents.head()
```

```python
df_contents.info()
```

```python
df_contents_clean = df_contents
revisiones_totales = df_contents.id.count()
```

```python
df_contents_clean["ediciones"] = df_contents_clean.groupby("revisor_username").transform("count")["revisor_id"]
```

```python
df_contents_clean["uno"] = df_contents_clean['ediciones'] == 1
df_contents_clean["+10"] = df_contents_clean['ediciones'] >= 10
df_contents_clean["+100"] = df_contents_clean['ediciones'] >= 100
```

```python
df_contents_clean
```

```python
df_contents_clean_no_nan = df_contents_clean[df_contents_clean["revisor_comment"].isna()]
df_contents_clean_no_nan
```

```python
comentarios_nulos_grupo1 = df_contents_clean_no_nan["uno"].sum()
comentarios_nulos_grupo10 = df_contents_clean_no_nan["+10"].sum()
comentarios_nulos_grupo100 = df_contents_clean_no_nan["+100"].sum()

porcentaje_grupo_1   = (comentarios_nulos_grupo1 *100) / revisiones_totales
porcentaje_grupo_10  = (comentarios_nulos_grupo10 *100) / revisiones_totales
porcentaje_grupo_100 = (comentarios_nulos_grupo100 *100) / revisiones_totales
```

```python
print("El porcentaje de contenidos que están publicados cuya última edición no tiene comentario para los usuarios que realizaron 1 de las últimas ediciones es {}%".format(porcentaje_grupo_1))
print("El porcentaje de contenidos que están publicados cuya última edición no tiene comentario para los usuarios que realizaron >10 de las últimas ediciones es {}%".format(porcentaje_grupo_10))
print("El porcentaje de contenidos que están publicados cuya última edición no tiene comentario para los usuarios que realizaron >100 de las últimas ediciones es {}%".format(porcentaje_grupo_100))
```
