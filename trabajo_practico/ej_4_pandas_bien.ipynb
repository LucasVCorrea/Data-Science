```python
df_contenidos = pd.read_csv("/content/drive/MyDrive/Organizacion De Datos/Trabajo_practico/contents.csv")
```

```python
df_languages.head()
```

```python
df_languages = df_languages[df_languages['babel_level'] != '0']
df_languages.head()
```

```python
df_contenidos.head()
```

```python
df_languages.rename(columns = {'babel_user':'revisor_id'}, inplace = True)
```

```python
nuevo_df = pd.merge(df_contenidos, df_languages, how="inner")
```

```python
nuevo_df.info()
```

```python
nuevo_df.babel_level.unique()
```

```python
nuevo_df
```

```python
agrupado = nuevo_df.groupby("babel_level").agg({"babel_level":["count"], "revisor_comment":["count"]})
```

```python
agrupado.columns = ["total_usuarios","con_comentarios"]
```

```python
agrupado["sin_comentario"] = (agrupado.total_usuarios - agrupado.con_comentarios)
```

```python
agrupado
```

```python
agrupado["tasa_revisiones_sin_comentario"] = (agrupado.sin_comentario / agrupado.total_usuarios) *100
```

la tasa de revisiones sin comentario que realizan en función de los idiomas que saben es la siguiente:




```python
agrupado[["tasa_revisiones_sin_comentario"]]
```
