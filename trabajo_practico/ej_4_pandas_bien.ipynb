```python
agrupado = nuevo_df.groupby("babel_level").agg({"babel_level":["count"], "revisor_comment":["count"]})
agrupado
```

```python
agrupado.columns = ["total_usuarios","con_comentarios"]
```

```python
agrupado["sin_comentario"] = (agrupado.total_usuarios - agrupado.con_comentarios)
```

```python
agrupado["tasa_revisiones_sin_comentario"] = (agrupado.sin_comentario / agrupado.total_usuarios) *100
```
