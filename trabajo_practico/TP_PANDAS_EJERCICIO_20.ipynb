```python
df_geo = pd.read_csv("/content/drive/MyDrive/Organizacion De Datos/Trabajo_practico/geo_tags.csv")
```

```python
df_geo.head(10)
```

```python
df_geo.info()
```

```python
df_geo_limpio = df_geo[['gt_page_id']]
df_geo_limpio
```

```python
agrupado = df_geo_limpio.groupby('gt_page_id').agg({'gt_page_id':["count"]})
```

```python
#df_limpio.loc[[1]].isna().sum().sum() # SUMA LA CANT DE VALORES NULOS
agrupado = agrupado.reset_index()
```

```python
agrupado.columns = ["id","references"]
agrupado
```

```python
df_mas_referencias = agrupado.nlargest(2,'references')
df_mas_referencias
```

```python
print("El id del segundo contenido con más referencias geográficas asignadas es {}".format(df_mas_referencias.iloc[1].id))
```
