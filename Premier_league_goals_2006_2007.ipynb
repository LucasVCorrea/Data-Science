```python
import pandas as pd
df_tabla = pd.read_csv("/content/results.csv")
```

```python
df_tabla
```

```python
df_tabla_2006_2007 = df_tabla[ df_tabla["season"] == '2006-2007' ] # --> me quedo solo con los del 2006 - 2007
df_tabla_2006_2007_limpio = df_tabla_2006_2007[["home_team", "away_team", "home_goals","away_goals"]]
```

```python
agrupado_local = df_tabla_2006_2007_limpio.groupby(["home_team"]).sum()
```

```python
agrupado_visitante = df_tabla_2006_2007_limpio.groupby(["away_team"]).sum()
```

Armo el Dataframe con los datos finales

```python
agrupado_local["away_goals"] = agrupado_visitante["away_goals"]
agrupado_local
```

Construyo el Dataframe final, con la suma total de los goles por equipo

```python
total = (agrupado_local["home_goals"]) + agrupado_local["away_goals"]
agrupado_local["total_goals"] = total
agrupado_local.reset_index()
```
