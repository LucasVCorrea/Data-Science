```python
import pandas as pd
df_raw = pd.read_csv("/content/NBA_season1718_salary.csv")
df_raw.head()
```

```python
df_raw.info()
```

Agrupo por equipos, para conseguir la suma total de los salarios y cual fué el mayor

```python
df_semi_raw = df_raw.groupby("Tm").agg({"season17_18":["sum","max"]})
df_semi_raw = df_semi_raw.reset_index()
df_semi_raw
```

Cambio el nombre de las columnas y agrego otra que indica el porcentaje del salario más caro contra el total del equipo (Cuanto se lleva el mejor pago respecto del total del equipo)

```python
df_semi_raw.columns = ["team","tot_paid","highest"]
df_semi_raw['porcentaje_h/tp'] = (( df_semi_raw["highest"]* 100) /df_semi_raw["tot_paid"])
#df_semi_raw.reset_index()
```

```python
df_semi_raw
```

Obtengo el equipo que más dinero gastó en salarios y cuanto ganó el mejor pago de la NBA en 2017 - 2018

```python
highest_paid_player = df_semi_raw["highest"].max() #Jugador más pago de toda la NBA 2017 - 2018
richest_team = df_semi_raw["tot_paid"].max() 
```

```python
import plotly.express as px
visual = px.scatter(df_semi_raw, x = "team", y = "tot_paid")
visual
```

```python
print("Salario más alto de toda la NBA 2017-2018 = ${} millones".format(highest_paid_player))
print("Total más alto del equipo que más gastó en la NBA 2017-2018 = ${} millones".format(richest_team))
```
