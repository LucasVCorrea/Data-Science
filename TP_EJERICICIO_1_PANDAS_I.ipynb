```python
df_logs_limpio = df_logs # copio para evitar el warning del .loc      
df_logs_limpio  
```

```python
df_logs_limpio["timestamp"] = pd.to_datetime(df_logs_limpio["timestamp"]).dt.date # paso la fecha a objeto fecha con datetime y dt.date para que me saque la hora
df_logs_limpio
```

```python
df_logs_limpio["actividades"] = 0   # agrego la columna actividades, para que tenga la cantidad que hubo
df_logs_limpio
```

```python
df_logs_limpio = df_logs_limpio[ ["timestamp","actividades"] ]  # me quedo con los datos que me interesan
df_logs_limpio
```

```python
agrupado_fechas = df_logs_limpio.groupby("timestamp").count()
#agrupado_fechas = agrupado_fechas.reset_index()
agrupado_fechas
```

```python
agrupado_fechas[ agrupado_fechas['actividades'] == agrupado_fechas.actividades.max()]
```

```python
agrupado_fechas[ agrupado_fechas['actividades'] == agrupado_fechas.actividades.min()]
```

```python
cant_actividad_minima = agrupado_fechas.actividades.min()
cant_actividad_maxima = agrupado_fechas.actividades.max()
maximo_dia = (agrupado_fechas[ agrupado_fechas['actividades'] == cant_actividad_maxima]).index[0]
minimo_dia = (agrupado_fechas[ agrupado_fechas['actividades'] == cant_actividad_minima]).index[0]
```

```python
print("1) Dia con más actividad:   {}   ({} acciones totales)".format(maximo_dia, cant_actividad_maxima))
print("2) Dia con menos actividad: {}   ({} acciones totales)".format(minimo_dia, cant_actividad_minima))
```
