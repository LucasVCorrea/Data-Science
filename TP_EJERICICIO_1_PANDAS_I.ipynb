```python
df_logs_limpio = df_logs # copio para evitar el warning del .loc      
```

```python
df_logs_limpio["timestamp"] = pd.to_datetime(df_logs_limpio["timestamp"]).dt.date # paso la fecha a objeto fecha con datetime y dt.date para que me saque la hora
df_logs_limpio
```

```python
# me quedo con los datos que me interesan
#df_logs_limpio.loc[:,"actividades"] = 0
df_logs_limpio = df_logs_limpio[ ["timestamp"] ]
```

```python
df_logs_limpio
```

```python
agrupado_fechas = df_logs_limpio.groupby("timestamp").agg({"timestamp":["count"]})
agrupado_fechas = agrupado_fechas.reset_index()
agrupado_fechas
```

```python
agrupado_fechas.columns = ["fecha","actividades"]
agrupado_fechas
```

```python
# filtrar con loc para ver que no se repita el maximo y el minimo (ya está hecho)
maxima_cantidad = agrupado_fechas.actividades.max()
minima_cantidad = agrupado_fechas.actividades.min()

agrupado_fechas["maximo"] = agrupado_fechas["actividades"] >= maxima_cantidad
agrupado_fechas["minimo"] = agrupado_fechas["actividades"] <= minima_cantidad
agrupado_fechas.head()
```

```python
agrupado_fechas[ agrupado_fechas["maximo"] ] #muestra las filas cuya actividad son mayores o iguales a la maxima
```

```python
agrupado_fechas[ agrupado_fechas["minimo"] ]    #muestra las filas cuya actividad son mayores o iguales a la minimia
```

```python
agrupado_fechas["fecha"][agrupado_fechas["maximo"]] #asi me muestra todas las fechas que tienen true en maximo (respuesta que me piden, ya está bien asi)
```

```python
agrupado_fechas["fecha"][agrupado_fechas["minimo"]] #asi me muestra todas las fechas que tienen true en minimo (respuesta que me piden, ya está bien asi)
```
