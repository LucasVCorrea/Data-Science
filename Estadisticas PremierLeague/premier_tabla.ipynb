```python
import pandas as pd
```

```python
df_premier_raw = pd.read_csv("/content/results.csv")
df_premier_raw
```

```python
df_premier_raw["result"].unique()
```

```python
def mapear_result_local(caracter):
  traduccion = {
      'H':3,
      'D':1,
      'A':0
  }

  return traduccion[caracter]
```

```python
def mapear_result_visit(caracter):
  traduccion = {
      'H':0,
      'D':1,
      'A':3
  }

  return traduccion[caracter]
```

```python
df_premier_raw["puntos_h"] = df_premier_raw["result"].map(mapear_result_local)
df_premier_raw["puntos_a"] = df_premier_raw["result"].map(mapear_result_visit)
```

```python
df_premier_semi_raw = df_premier_raw[df_premier_raw["season"] == "2017-2018"]
df_premier_semi_raw
```

```python
df_premier_clean = df_premier_semi_raw[ ["home_team","result","season","puntos_h"] ]
df_premier_clean_2 = df_premier_semi_raw[ ["away_team","result","season","puntos_a"] ]
```

```python
agrupado_local = df_premier_clean.groupby(["home_team"]).sum()
```

```python
agrupado_visit = df_premier_clean_2.groupby("away_team").sum()
```

```python
agrupado_local["puntos_a"] = agrupado_visit["puntos_a"]
agrupado_local
```

```python
total = agrupado_local["puntos_h"] + agrupado_visit["puntos_a"]
agrupado_local["points"] = total
agrupado_local
```

```python
table_2017_2018_ = agrupado_local.sort_values(by = ["points"], ascending = False)
table_2017_2018_
```
